### Load Dataset

```python
import pandas as pd
dataset = pd.read_csv('data/LinkedIn Job Search.csv')
dataset
```

### Drop Kolom yang Tidak Relevan

```python
dataset = dataset.drop(columns=['Unnamed: 15', 'Unnamed: 16'])
dataset
```

### Exploratory Data Analysis

```python
# Melihat informasi umum tentang dataset
dataset.info()
```

```python
# Melihat jumlah nilai yang hilang di setiap kolom
dataset.isnull().sum()
```

```python
# Menghapus missing value
dataset = dataset.dropna()
dataset.isna().sum()
```

```python
# Mengecek duplikasi dalam dataset
dataset.duplicated().sum()
```

```python
# Menghapus duplikasi
dataset = dataset.drop_duplicates()
dataset.duplicated().sum()
```

### Feature Selection

```python
# Memilih fitur yang relevan untuk sistem rekomendasi
selected_features = ['Job Title', 'Job Qualifications', 'Type', 'Work Hours', 'URL']
data_cleaned = dataset[selected_features]
data_cleaned
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = data_cleaned.copy()

# 3. Gabungkan semua fitur menjadi satu string
df['combined_features'] = df['Job Title'] + ' ' + df['Job Qualifications'] + ' ' + df['Type'] + ' ' + df['Work Hours']

# 4. Vektorisasi teks menggunakan TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['combined_features'])

# 5. Input dari user
user_input = 'Data Scientist Machine Learning Remote Full-time'
user_vector = tfidf.transform([user_input])

# 6. Hitung cosine similarity
similarities = cosine_similarity(user_vector, tfidf_matrix)
similarities

# 7. Ambil Top-N rekomendasi
top_n = 2
top_indices = similarities[0].argsort()[-top_n:][::-1]
recommendations = df.iloc[top_indices]

# 8. Tampilkan hasil rekomendasi
print("Rekomendasi pekerjaan untuk input pengguna:")
for idx, row in recommendations.iterrows():
    print(f"- {row['Job Title']} ({row['Type']}, {row['Work Hours']})")
    print(f"  Kualifikasi: {row['Job Qualifications']}")
    print(f"  Lokasi: {row['URL']}")
    print()
```
