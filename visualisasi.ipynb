```python
import folium 
import requests
import json
import pandas as pd
from folium.plugins import FastMarkerCluster
```

```python
dataset = pd.read_csv("data/Linkedin Job Search.csv")
dataset = dataset.drop(columns=["Unnamed: 15","Unnamed: 16"])
kota_list = ['Jakarta', 'Surabaya', 'Bandung', 'Cikarang', 'Denpasar', 'Batang', 'Yogyakarta', 'Malang',
              'Pontianak', 'Kendal', 'Muarajawa', 'Tangerang', 'Pekanbaru', 'Medan', 'Lampung', 'Kuta',
              'Ubud', 'Bali','Bekasi', "East Java"]

# Fungsi untuk mengganti lokasi dengan nama kota sesuai daftar kota
def ganti_kota(lokasi):
    if isinstance(lokasi, str):  # Memastikan bahwa nilai adalah string
        for kota in kota_list:
            if kota in lokasi:
                return kota
    return lokasi

df = dataset.copy()
# Mengganti lokasi yang mengandung kota dengan nama kota yang sesuai
df['Company Location'] = df['Company Location'].apply(ganti_kota)
df['Company Location'] = df["Company Location"].replace(to_replace="East Java", value="Jawa Timur")
df['Company Location'] = df["Company Location"].replace(to_replace="Lampung", value="Bandar Lampung")
df = pd.DataFrame(df)
df
```

```python
with open("data/location.json",'r') as f:
    data = json.load(f)
 
data_geo = pd.DataFrame(data)
print(data_geo.head(5))
```

```python
df_jobs = df.merge(data_geo, left_on='Company Location', right_on='Kota', how='left')
df_jobs = pd.DataFrame(df_jobs)
df_jobs = df_jobs.drop(columns=["Company Location"], axis=1)
df_jobs = df_jobs.dropna(subset=["Kota"])
display(df_jobs) 
```

```python
records = []
keys = ("Job Title","Job Qualifications","URL","Type","Work Hours","Kota","Georeferenced")

for index, row in df_jobs.iterrows():
    records.append({key: row[key] for key in keys})

print(records[0])
```

```python
for record in records:
    coords = record["Georeferenced"]
    longitude, latitude = coords.split("(")[-1].split(")")[0].split()
    record["Longitude"] = float(longitude) 
    record["Latitude"] = float(latitude) 

records[4]
```

```python
records = pd.DataFrame(records)
records
```

```python
records.to_csv("data/data.csv", index=False)
```

```python
import pandas as pd
import re

# Membuat fungsi untuk mengekstrak inti dari nama pekerjaan
def extract_job_title(title):
    # Menghapus konten dalam tanda kurung dan menghilangkan spasi di awal/akhir
    title = re.sub(r'\(.*?\)', '', title).strip()
    # Memisahkan dengan koma dan mengambil kata pertama
    title = title.split(',')[0]
    # Menghapus kata-kata tambahan umum seperti "Senior", "Junior", dll.
    title = re.sub(r'\b(Senior|Junior|Lead|Head|Manager|Specialist|Full-time|Consultant|Developer|Instructor|Lecturer|GoTo Financial|Gojek|dyeing|in|MarTech|Indonesia|Intern|Experimentation)\b', '', title, flags=re.IGNORECASE).strip()
    # Menghapus tanda "-"
    title = title.replace('-', '').strip()
    return title

# Menerapkan fungsi ke kolom Job Title
df['Job Title'] = df['Job Title'].apply(extract_job_title)
df.head(10)
```

```python
# Membuat daftar kata kunci yang ingin dipertahankan
keywords = ['Data', 'Analyst', 'Analys', 'Analis', 'Analisa', 'Analisis', 
            'Analitik', 'Sains', 'Science', 'Scientist', 'Sciences', 
            'Engineer', 'Machine', 'Learning', 'Mesin', 'AI', 'Artificial', 'Intelligence']

# Membuat regex pattern dari daftar kata kunci
pattern = '|'.join(keywords)

# Menyaring baris yang mengandung salah satu kata kunci
df = df[df['Job Title'].str.contains(pattern, flags=re.IGNORECASE, na=False)]
# df = df.drop(columns=["Unnamed: 0"])

# Mengatur ulang indeks agar tidak mengikuti data lama
df = df.reset_index(drop=True)
df.head(10)
```

```python
import folium
from folium.plugins import MarkerCluster

# Membuat peta dengan posisi awal Indonesia
map = folium.Map(location=[-0.789275, 113.921327], zoom_start=5)

# Membuat MarkerCluster
marker_cluster = MarkerCluster().add_to(map)

# Menambahkan marker dengan cluster
for index, record in df.iterrows():
    pekerjaan_table = """
        <table style='width:100%; table-layout:fixed;'>
            <tr><th style='white-space:nowrap;'>Pekerjaan</th></tr>
            <tr><td style='white-space:normal; word-wrap:break-word;'>{}</td></tr>
        </table>
    """.format(record["Job Title"])
    
    popup_content = f"<h3>{record['Kota']}</h3>{pekerjaan_table}"

    folium.Marker(
        [record["Latitude"], record["Longitude"]],
        popup=folium.Popup(popup_content, max_width=250),
        tooltip=f"Lowongan di {record['Kota']}",
        icon=folium.Icon(icon="info-sign", color="blue")
    ).add_to(marker_cluster)

# Menampilkan peta
map
```
